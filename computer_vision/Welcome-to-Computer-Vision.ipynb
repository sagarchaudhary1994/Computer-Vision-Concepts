## Computer Vision Pipeline

Steps in Computer Vision Pipeline are:
* Input Data >> Image Pre-Processing  >> Feature Extraction >> ML Model


* Computer Vision Pipeline for simple image classification


**How does this pipeline work?**
- Input Image- It can be image or frames from video
- Image Preprocessing - Purpose is to standardize the image. Steps can include, image resizing, bluring, rotation, changing its shape, color transformation from colored image to grayscale. Only by standardization of images like resizing to same shape, images can then be compared and analyzed.
- Feature Extraction- Features are generally the information helping in identifing the object say color, object shape. This step returns a feature vector, which is a list of all the unique features (or say unique shapes) for the object. Ex. in this bike example features can be wheel, handle, mudguard.
- Model- Feature vector obtained in feature extraction step is then feed to image classification model, which predicts the class of the image. The output for model is the probalities for each class.
